```python
sentence = '''
    Dr. A.P.J. Abdul Kalam was a great scientist and former President of India. He inspired millions of students through his speeches, books, and motivational thoughts. He believed that hard work, dedication, and dreaming big could help every person achieve success in life.
'''
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
list_words = word_tokenize(sentence)
print(list_words)
```

```python
words = nltk.pos_tag(list_words)
```
