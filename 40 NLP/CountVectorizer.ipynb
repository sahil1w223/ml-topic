```python
df = pd.read_csv('/content/SMSSpamCollection.txt',sep='\t',names=['output','messages'])
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize
import re
```

```python
corpus = []
for i in range(len(df['messages'])):
  df['messages'][i] =  re.sub('[^a-zA-Z]',' ',df['messages'][i])
  df['messages'][i] = df['messages'][i].lower()
  df['messages'][i] = df['messages'][i].split()
  lemmatizer = WordNetLemmatizer()
  df['messages'][i] = [lemmatizer.lemmatize(word) for word in df['messages'][i] if word not in set(stopwords.words('english'))]
  df['messages'][i] = ' '.join(df['messages'][i])
  corpus.append(df['messages'][i])
```

```python
corpus
```

```python
y = pd.get_dummies(df['output'])
```

```python
y = y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(corpus,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

x_train_transformed = cv.transform(x_train).toarray()
x_test_transformed = cv.transform(x_test).toarray()

spam_detect_model = MultinomialNB().fit(x_train_transformed,y_train)
```

```python
y_pred = spam_detect_model.predict(x_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
