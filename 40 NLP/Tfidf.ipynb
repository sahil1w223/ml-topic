```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('/content/SMSSpamCollection.txt', sep='\t', names=['Output', 'Text'])
df.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
words = stopwords.words('english')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
stemms = WordNetLemmatizer()
```

```python
corous = []
for i in range(len(df['Text'])):
  df['Text'][i] = re.sub('[^a-zA-Z]', ' ', df['Text'][i])
  df['Text'][i] = (df['Text'][i]).lower()
  df['Text'][i] = df['Text'][i].split()
  df['Text'][i] = [stemms.lemmatize(word) for word in df['Text'][i] if word not in set(words)]
  df['Text'][i] = ' '.join(df['Text'][i])
  corous.append(df['Text'][i])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
x = TfidfVectorizer().fit_transform(corous).toarray()
y = pd.get_dummies(df['Output'])
y = y.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
cv = MultinomialNB()
cv.fit(x_train, y_train)
y_pred = cv.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
