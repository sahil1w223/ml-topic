```python
from nltk.stem import PorterStemmer,RegexpStemmer
```

```python
words = ['eating','eats','eaten','writing', 'writes', 'programing', 'programs', 'history','finally','fimalized']
words
```

```python
stemming = PorterStemmer()
for w in words:
    print(f"{w}->{stemming.stem(w)}")
```

```python
reg_stemming = RegexpStemmer('ing$|s$|e$|able$|en', min=4)
for w in words:
    print(f"{w}->{reg_stemming.stem(w)}")
```

```python
from nltk.stem import SnowballStemmer
stemming = SnowballStemmer(language='english')
for w in words:
    print(f"{w}->{stemming.stem(w)}")
```

# difference between PorterStemmer and SnowballStemmer

```python
PorterStemmer().stem('fairly'),PorterStemmer().stem('sportingly')
```

```python
SnowballStemmer(language='english').stem('fairly'),SnowballStemmer(language='english').stem('sportingly')
```

```python
from nltk.stem import WordNetLemmatizer

stemming = WordNetLemmatizer()
for w in words:
    print(f'{w}->{stemming.lemmatize(w,pos='v')}')
```

# difference Between Stemming and Lemmatization

| -------------------- | ---------------------- |
| Word cut karta hai   | Meaning samajhta hai   |
| Fast                 | Thoda slow             |
| studies → studi ❌    | studies → study ✅      |
