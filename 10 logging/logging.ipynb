What is Logging in Python?

Logging means recording information/messages about what your program is doing.
“Logging is the process of recording events, errors, and messages while a program is running.”

```python
import logging

#  basic configuration for logging
logging.basicConfig(level=logging.ERROR,format='%(asctime)s - %(levelname)s - %(message)s',force=True, filename='app.log')  # Set the logging level to INFO

print(logging.getLogger().level)  # This will print the current logging level
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

Python me 5 main logging levels hote hain.

Inko low priority se high priority tak samjho 👇

| Level      | Number | Meaning                                  |
| ---------- | -----: | ---------------------------------------- |
| `DEBUG`    |     10 | Detailed information (debugging ke liye) |
| `INFO`     |     20 | Normal information                       |
| `WARNING`  |     30 | Warning message                          |
| `ERROR`    |     40 | Error hua                                |
| `CRITICAL` |     50 | Very serious error                       |
