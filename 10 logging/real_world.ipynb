```python
import logging
logging.basicConfig(level=logging.DEBUG,format='%(asctime)s - %(levelname)s - %(message)s',force=True,handlers=[logging.FileHandler('app.log'), logging.StreamHandler()])
```

```python
logger = logging.getLogger('ArticleLogger')

def sum(a,b):
    result = a + b
    logger.debug(f'Adding {a} and {b}: {result}')
    return a + b

def subtract(a,b):
    result = a - b
    logger.debug(f'Subtracting {b} from {a}: {result}')
    return a - b

def multiply(a,b):
    result = a * b
    logger.debug(f'Multiplying {a} and {b}: {result}')
    return a * b

def divide(a,b):
    try:
        result = a / b
        logger.debug(f'Dividing {a} by {b}: {result}')
        return result
    except ZeroDivisionError:
        logger.error('Division by zero is not allowed')
        return None
    
```
